```python
g = sns.violinplot(y='Species', x='SepalLengthCm', data=df, inner='quartile')
plt.show()
g = sns.violinplot(y='Species', x='SepalWidthCm', data=df, inner='quartile')
plt.show()
g = sns.violinplot(y='Species', x='PetalLengthCm', data=df, inner='quartile')
plt.show()
g = sns.violinplot(y='Species', x='PetalWidthCm', data=df, inner='quartile')
plt.show()
```

```python
df.boxplot(by="Species", figsize=(10, 10));
```

```python
sns.FacetGrid(df, hue="Species").map(sns.distplot,"PetalLengthCm").add_legend()
sns.FacetGrid(df, hue="Species").map(sns.distplot,"PetalWidthCm").add_legend()
sns.FacetGrid(df, hue="Species").map(sns.distplot,"SepalLengthCm").add_legend()
plt.show();
```
