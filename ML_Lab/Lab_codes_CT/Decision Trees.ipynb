# Preprocess

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Age'] = le.fit_transform(df['Age'])
df['Income'] = le.fit_transform(df['Income'])
df['Gender'] = le.fit_transform(df['Gender'])
df['Martial Status'] = le.fit_transform(df['Martial Status'])
df['Buys'] = le.fit_transform(df['Buys'])
df
```

# Create Decision Tree

## Split data

```python
from sklearn.model_selection import train_test_split

X_train , X_test, y_train, y_test = train_test_split(df.iloc[:,1:-1], df.iloc[:,-1], test_size=0.2, random_state=5)
print(len(X_train))
print(len(X_test))
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
from six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(model, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

# Default Dataset

## Readfile

```python
from sklearn import datasets
import pandas as pd
wine = datasets.load_wine()
X = pd.DataFrame(data=wine.data, columns=wine.feature_names)
y = pd.DataFrame(wine.target, columns=['Target'])
X.head()
```

```python
y.head()
```

# Create Decision Tree

## Split data

```python
from sklearn.model_selection import train_test_split

X_train , X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=5)
print(len(X_train))
print(len(X_test))
X_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

## Accuracy

```python
model.score(X_test, y_test)
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, model.predict(X_test))
```

## Plot Decision Tree

```python
from six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(model, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

## Cross-Validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, wine.data, wine.target, cv=5)
scores
```
